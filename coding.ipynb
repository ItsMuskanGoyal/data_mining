```python
import pandas as pd
```

```python
df = pd.read_csv('salaries_dataset.csv')
df.head()
```

```python
df.drop(columns=['employment_type', 'salary', 'salary_currency', 'employee_residence', 'remote_ratio', 'company_location'], inplace=True)
```

```python
df.describe()
```

```python
df.head()
```

```python
df['salary_greater_than_200k'] = df['salary_in_usd'].apply(lambda x : 1 if x >= 200000 else 0)
```

```python
df.head()
```

```python
inputs = df[['work_year','experience_level', 'job_title', 'company_size']]
target = df['salary_greater_than_200k']
```

```python
inputs.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
lb_company = LabelEncoder()
lb_job = LabelEncoder()
lb_ex = LabelEncoder()

inputs['experience_level']= lb_ex.fit_transform(inputs['experience_level']) 
inputs['job_title']= lb_job.fit_transform(inputs['job_title']) 
inputs['company_size']= lb_company.fit_transform(inputs['company_size']) 
inputs.head()
```

```python
inputs.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2)
```

```python
inputs.describe()
```

```python
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

This will predict whether that perosn will make more than 200k or not
After giving these parameters 
work_year	experience_level	job_title	company_size
